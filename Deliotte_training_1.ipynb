```python
lower, upper, special, digit = 0, 0, 0, 0
password = input("Enter your password")
if (len(password) >= 8):
    for i in password:
        for word in password.split():
            if(word[0].isupper()):
                upper += 1
        if(i.islower()):
            lower += 1
        if(i.isdigit()):
            digit += 1
        if(i == '@' or i == '$' or i == '_' or i == '#'):
            special += 1
else:
    print("Password should be more than 6 characters")
if (lower >= 1 and upper >= 1 and special >= 1 and digit >= 1):
    print("Valid Password")
else:
    print("Invalid Password")
```

```python
Password=input()
Upper_case,Lower_case,Digit=0,0,0
for char in Password:
    if char.isupper():
        Upper_case=Upper_case+1
    elif char.islower():
        Lower_case=Lower_case+1
    elif char.isdigit():
        Digit=Digit+1
if (Upper_case>=1 and Lower_case>=1 and Digit>=1 and (Upper_case+Lower_case+Digit)>=8):
    print("VALID")
else:
    print("IN VALID")
```
